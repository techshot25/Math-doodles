### This is to find the value of some definite integral using complex analysis.
We want to find the value of the integral
$$\int_{0}^{2\pi}e^{\cos{x}}\; \cos{\left ( \sin{x} \right ) }\; dx$$

The trick will be to append another integral that looks similar to the original one with a few differences.

$$\int_{0}^{2\pi}e^{\cos{x}}\; \cos{\left ( \sin{x} \right ) }\; dx + i \int_{0}^{2\pi}e^{\cos{x}}\; \sin{\left ( \sin{x} \right ) }\; dx$$

This allows us to use Euler's formula:
$$e^{i \theta}=\cos{\theta}+i\sin{\theta}$$
Now we can write the integal like this:
$$\int_{0}^{2\pi}e^{\cos{x}}\: e^{\left (i \sin{x} \right ) }\; dx = \int_{0}^{2\pi}e^{\left (\cos{x}+i \sin{x} \right ) }\; dx = \int_{0}^{2\pi} e^{e^{ix} }\; dx$$

This last integral is a bit more complicated to breakdown, so we can refer to the series expansion of $e^{x}$ at $x=0$.

$$\int_{0}^{2\pi} \sum_{n=0}^{\infty} \frac{\left ( e^{ix} \right )^{n}}{n!} \; dx$$

We need to be careful because when $n=0$ we have a bit of a problem if we use the fundamental theorem of calculus.

$$\int_{0}^{2\pi} dx + \int_{0}^{2\pi} \sum_{n=1}^{\infty} \frac{\left ( e^{ix} \right )^{n}}{n!} \; dx = 2\pi+ \int_{0}^{2\pi} \sum_{n=1}^{\infty} \frac{ e^{inx}}{n!} \; dx$$

Now we know the first integral breaks down to $2\pi$ and the second goes like this:

$$2\pi + \left[ \sum_{n=1}^{\infty} \frac{ e^{inx} }{n! \left( i\,n \right)} \right]_{x=0}^{x=2\pi}$$

But we know that $e^{0} = 1$ and $e^{2n\pi i} = \cos{\left(2n\pi\right)} + i \sin{\left(2n\pi\right)} = 1\;\; \text{so} \;\; e^{2\pi n}-e^{0}=0 \;\; \forall \; n \in \mathbb{Z}\;$ and the only term that survives is $2\pi$. 

Also considering the following definition of a complex integrand:

If $\;\;\int f(x)\,dx+i\int g(x)\,dx = u(x)+i\,v(x)\;\;$ then $\;\;\int f(x)\,dx = u(x)\;\;$ and $\;\;\int g(x)\,dx = v(x)\;\;$ given that $\;\;f,g,u,v \in \mathbb{R}$.


So in the end our answer becomes:

$$\boxed{\int_{0}^{2\pi}e^{\cos{x}}\; \cos{\left ( \sin{x} \right ) }\; dx=2\pi}$$

```python
from math import *

def f(x):
    return exp(cos(x))*cos(sin(x))

dx = 1.0e-5
a = 0 
b = 2*pi

x = a
ans = 0
while x < b:
    ans += f(x)*dx
    x += dx

print("The integral gives us around {:.5} using a left endpoint sum. And 2*pi is around {:.5}".format(ans,2*pi))
```
